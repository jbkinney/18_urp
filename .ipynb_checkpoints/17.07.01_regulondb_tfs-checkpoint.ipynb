```python
# Always put this first
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pdb
```

```python
# Open file and load all lines into list
db_local_file = 'binding_site_db.txt'
f = open(db_local_file)
lines = f.readlines()
print len(lines)
lines[-1]
```

```python
# Count the number of (equal length) binding sites for each TF

# Start by listing all binding sites for each TF
tf_sites_dict = {}
for line in lines:
    if line[0]=='#':
        continue
    atoms = line.split('\t')
    if len(atoms) < 12:
        continue
    tf = atoms[1]
    site = atoms[11]
    if not tf in tf_sites_dict:
        tf_sites_dict[tf] = []
    tf_sites_dict[tf].append(site)
    
    
# Pick optimal length site for each TF
tf_names = tf_sites_dict.keys()
for tf in tf_names:
    #print tf
    sites = tf_sites_dict[tf]
    site_lengths = np.array([len(s) for s in sites if len(s) > 0])
    if len(site_lengths) == 0:
        del tf_sites_dict[tf]
    else:
        length_dict = {}
        for l in site_lengths:
            if l in length_dict.keys():
                length_dict[l] += 1
            else:
                length_dict[l] = 1
        count = length_dict.values()
        lengths = length_dict.keys()
        most_common_length = lengths[np.argmax(count)]
        sites_to_keep = [s.upper() for s in sites if len(s)==most_common_length]
        tf_sites_dict[tf] = sites_to_keep
        #print 'tf: %s, lenght: %d, count: %d'%(tf,most_common_length,len(sites_to_keep))
        
# Sort TFs by their length
tf_names = tf_sites_dict.keys()
df = pd.DataFrame(columns=['tf','num_sites','site_length'])
df.loc[:,'tf'] = tf_names
df.loc[:,'num_sites'] = [len(tf_sites_dict[tf]) for tf in tf_names]
df.loc[:,'site_length'] = [len(tf_sites_dict[tf][0]) for tf in tf_names]
df.sort_values(by='num_sites',ascending=False,inplace=True)
df.set_index('tf',drop=True,inplace=True)
```

```python
def get_sites(tf,start=0,stop=None):
    raw_sites = tf_sites_dict[tf]
    if stop is None:
        stop = len(raw_sites[0])
    sites = [s[start:stop] for s in raw_sites]
    return sites
```

```python
# Create weight matrix from set of sites
def get_ct_profile(tf,start=0,stop=None):
    sites = get_sites(tf,start=start,stop=stop)
    # Get site length
    assert len(sites) > 0
    L = len(sites[0])
    
    # Create PWM
    bases = 'ACGT'
    ct_profile = np.zeros([L,4])
    for site in sites:
        assert len(site)==L
        
        chars = np.array([c for c in site.upper()])
        for i,b in enumerate(bases):
            ct_profile[:,i] += chars==b
        
    return ct_profile
```

```python
# Get information profile of site
def get_info(tf,pseudocount=0,start=0,stop=None):
    cts = get_ct_profile(tf,start=start,stop=stop)
    cts += pseudocount
    ps = cts/cts.sum(axis=1)[:,np.newaxis]
    info = 2-(-ps*np.log2(ps+1E-6)).sum(axis=1)
    return info
```

```python
# Get optimal window of given length
def get_best_window(tf,window_length,pseudocount=1):
    info = get_info(tf,pseudocount=pseudocount,start=0,stop=None)
    site_length = len(info)
    if site_length < window_length:
        return (0,site_length)
    window_info_totals = []
    for k in range(site_length-window_length+1):
        window_info_totals.append(sum(info[k:(k+window_length)]))
    
    start = np.array(window_info_totals).argmax()
    stop = start + window_length 
    return (start,stop)
```

```python
def get_consensus(tf,start=0,stop=None):
    cts = get_ct_profile(tf,start=start,stop=stop)
    L = cts.shape[0]
    bases = 'ACGT'
    consensus = ''
    for k in range(L):
        i = np.argmax(cts[k,:])
        consensus += bases[i]
    return consensus
```

```python
def get_pwm(tf,start=0,stop=None,pseudocount=1):
    cts = get_ct_profile(tf,start=start,stop=stop)
    cts += pseudocount
    ps = cts/cts.sum(axis=1)[:,np.newaxis]
    pwm = np.log(ps)
    return pwm
```

```python
def evaluate_pwm(pwm,seq,subtract_mean=True):  
    L = len(seq)
    assert pwm.shape==(L,4)
    
    # Compute consensus score
    consensus_score = 0
    bases = 'ACGT'
    score = 0
    for k, c in enumerate(seq):
        for i,b in enumerate(bases):
            score += pwm[k,i]*(b==c)
            
    if subtract_mean:
        score -= pwm.mean(axis=1).sum()
        
    return score
```

```python
# Fill in information for each tf
window_length=30

df.loc[:,'num_sites'] = [len(tf_sites_dict[tf]) for tf in df.index]
starts=[]
stops=[]
infos=[]
consensus_seqs=[]
consensus_scores=[]
for tf in df.index:
    start,stop = get_best_window(tf,window_length)
    info_profile = get_info(tf,start=start,stop=stop,pseudocount=1)
    pwm = get_pwm(tf,start=start,stop=stop,pseudocount=1)
    consensus_seq = get_consensus(tf,start=start,stop=stop)
    #pdb.set_trace()
    consensus_score = evaluate_pwm(pwm,consensus_seq)
    
        
    # Record stuff
    starts.append(start)
    stops.append(stop)
    infos.append(info_profile.sum())
    consensus_seqs.append(consensus_seq)
    consensus_scores.append(consensus_score)
    
df.loc[:,'start']=starts
df.loc[:,'stop']=stops
df.loc[:,'info']=infos
df.loc[:,'cons_seq']=consensus_seqs
df.loc[:,'cons_score']=consensus_scores
```

```python
# Show TFs with sufficient information in window
info_threshold = 8.0
num_sites_threshod = 10
df_ok = df[(df['info'] >= info_threshold) & (df['num_sites'] >= num_sites_threshod)]
print 'num_sites == %d'%len(df_ok)
df_ok
```

```python
tfs = df_ok.index
num_tfs = len(tfs)
plt.figure(figsize=[16,.6*num_tfs])
for i,tf in enumerate(tfs):
    start=df_ok.loc[tf,'start']
    stop=df_ok.loc[tf,'stop']
    num_sites=df_ok.loc[tf,'num_sites']
    
    # Get and show count profile
    cts = get_ct_profile(tf,start=start,stop=stop)
    plt.subplot(num_tfs/2,4,2*i+1)
    plt.imshow(cts.T,interpolation='nearest')
    plt.title('%s:%d sites'%(tf,num_sites))
    plt.yticks(range(4),'ACGT')
    plt.clim([0,max(np.ravel(cts))])
    
    # Get and show info
    info = get_info(tf,pseudocount=1,start=start,stop=stop)
    total_info = info.sum()
    plt.subplot(num_tfs/2,4,2*i+2)
    plt.bar(range(len(info)),info)
    plt.ylim([0,2])
    plt.title('%s: %.1f bits'%(tf,total_info))

plt.tight_layout()
plt.savefig('candidate_tfs.png')
plt.show()
```
